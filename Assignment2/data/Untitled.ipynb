```python
from csv import reader
from matplotlib import pyplot as plt
```

```python
dataset = list()
with open('Assignment2_river_data.csv', 'r') as file:
        csv_reader = reader(file)
        for row in csv_reader:
            if not row:
                continue
            dataset.append(row)
dataset.pop(0)
dataset
```

```python
for i in range(len(dataset)):
    plt.scatter(dataset[i][0],dataset[i][1],marker='.')
plt.show
```
